#**Matplotlib**

Matplotlib is one of the most popular Python packages used for data visualization. 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

Creating Datafame

```python
# create 2D array of table given above 
data = [['E001', 'M', 34, 123, 'Normal', 350], 
        ['E002', 'F', 40, 114, 'Overweight', 450], 
        ['E003', 'F', 37, 135, 'Obesity', 169], 
        ['E004', 'M', 30, 139, 'Underweight', 189], 
        ['E005', 'F', 44, 117, 'Underweight', 183], 
        ['E006', 'M', 36, 121, 'Normal', 80], 
        ['E007', 'M', 32, 133, 'Obesity', 166], 
        ['E008', 'F', 26, 140, 'Normal', 120], 
        ['E009', 'M', 32, 133, 'Normal', 75], 
        ['E010', 'M', 36, 133, 'Underweight', 40] ] 
  
# dataframe created with 
# the above data array 
df = pd.DataFrame(data, columns = ['EMPID', 'Gender',  
                                    'Age', 'Sales', 
                                    'BMI', 'Income'] ) 
  
print(df)
```

```python
print(np.shape(df))
```

```python
# create histogram for numeric data 
df.hist() 
  
# show plot 
plt.show() 
```

```python
# Plot the bar chart for numeric values 
# a comparison will be shown between 
# all 3 age, income, sales 
df.plot.bar()
```

```python
#plot between 2 attributes 
plt.bar(df['Age'], df['Sales']) 
plt.xlabel("Age") 
plt.ylabel("Sales") 
plt.show() 
```

#### Make data using numpy

```python
x = np.linspace(1, 10, 10) # make data using numpy np.linspace(start, stop, num)
y = x**2
z = x**2+2
print("x", x)
print("y", y)
print("z", z)
```

```python
plt.plot(x, y, 'green') # plot a line using one line
```

```python
plt.subplot(1, 2, 1)  # plot subgraph using subplots(rows, columns, indexes)
plt.plot(x, y, 'r--') # color and line style 
plt.title("a")
plt.subplot(1, 2, 2)
plt.plot(x, z, 'g*--')
plt.title("b")
```

```python
plt.subplot(2, 1, 1)  # plot subgraph using subplots(rows, columns, indexes)
plt.plot(x, y, 'rs') # third value takes color and line style , s Represents square markers
plt.subplot(2, 1, 2)
plt.plot(x, z, 'g--v') # v Represents triangle-down markers
```

```python
fig, axes = plt.subplots(dpi=50)
axes.set_xlabel("x-axis") 
axes.set_ylabel("y-axis") 
axes.set_title("title of the graph")
axes.plot(x**2, 'r')
axes.plot(x*3, 'g')
axes.legend(["label-1","label-2"], loc=0) #Makes legend of label-1 and label-2 in the graph
```

```python
fig, axes = plt.subplots(dpi=100)
# colors in matplotlib
axes.plot(x, x+1, color='red')
axes.plot(x, x+2, color='green')
axes.plot(x, x+3, color='blue')
```

```python
fig, axes = plt.subplots(dpi=100)
# transparency of colors in matplotlib
axes.plot(x, x+1, color='red', alpha=1)
axes.plot(x, x+2, color='green', alpha=0.5)
axes.plot(x, x+3, color='blue', alpha=0.3)
```

```python
fig, axes = plt.subplots(dpi=100)
# linwidth and linestyles
axes.plot(x, x+1, color='red', linewidth=1, linestyle='--')
axes.plot(x, x+2, color='green', linewidth=10, linestyle='-.')
axes.plot(x, x+3, color='blue', linewidth=5, linestyle=':')
```

```python
fig, axes = plt.subplots(dpi=100)
# markers on line
axes.plot(x, x+1, color='red', marker='o')
axes.plot(x, x+2, color='green', marker='+')
axes.plot(x, x+3, color='blue', marker='s')
```

```python
fig, axes = plt.subplots(dpi=100)
# size of markers
axes.plot(x, x+1, color='green', marker='+', markersize=10)
axes.plot(x, x+2, color='green', marker='s', markersize=10)
axes.plot(x, x+3, color='green', marker='o', markersize=10)
```

```python
fig, axes = plt.subplots(dpi=100)
# markerfacecolor
axes.plot(x, x+1, color='green', marker='+', markersize=10, markerfacecolor='red')
axes.plot(x, x+2, color='green', marker='s', markersize=10, markerfacecolor='blue')
axes.plot(x, x+3, color='green', marker='o', markersize=10, markerfacecolor='red')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].plot(x, x**2, x, x**3, linewidth=2)
axes[1].plot(x, x**2, x, x**3, linewidth=2)
axes[1].set_ylim([1, 60])
axes[1].set_xlim([1, 5])
```

2D plots

```python
n = np.array([0, 1, 2, 3, 4, 5])
print(len(n))
fig, axes = plt.subplots(1, 4, figsize=(15,5))
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15,5))
axes[0].set_title("scatter plot")
axes[0].scatter(n, n+0.25*np.random.randn(len(n)), color='blue')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15,5))

axes[0].set_title("scatter plot")
axes[0].scatter(n, n+0.25*np.random.randn(len(n)), color='blue')

axes[1].set_title("step plot")
axes[1].step(n, n**2, linewidth=2, color='blue')
```

```python
n = np.array([1, 2, 3, 4, 5, 6])
fig, axes = plt.subplots(1, 4, figsize=(15,5))

axes[0].set_title("scatter plot")
axes[0].scatter(n, n+0.25*np.random.randn(len(n)), color='blue')

axes[1].set_title("step plot")
axes[1].step(n, n**2, linewidth=2, color='blue')

axes[2].set_title("bar graph")
axes[2].bar(n, n**2, align="center", width=0.5, color='blue')

axes[3].set_title("fill area between curves")
axes[3].fill_between(n, n**2, n**3, color='blue')
```
